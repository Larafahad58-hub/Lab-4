```python
# اختيار العمود الذي ترغب في العمل عليه (في هذه الحالة "Fare" كمثال)
column_name = 'Fare'  # استبدل "Fare" بالعمود الذي تود العمل عليه

# التأكد من أن العمود موجود
if column_name in df.columns:
    # حساب الربيع الأول (Q1) والربيع الثالث (Q3)
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    # حساب القيم الدنيا والعليا
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # حذف القيم الشاذة (إزالة الصفوف التي تحتوي على القيم الشاذة)
    df_cleaned = df[(df[column_name] >= lower) & (df[column_name] <= upper)]
    
    # عرض أول 15 صف من البيانات بعد حذف القيم الشاذة
    print(df_cleaned.head(15))
else:
    print(f"العمود '{column_name}' غير موجود في البيانات.")
```

```python
# اختيار العمود الذي ترغب في العمل عليه (مثال "Fare" أو "Age")
column_name = 'Fare'  # استبدل "Fare" بالعمود الذي تود العمل عليه مثل 'Age' أو 'SibSp'

# التأكد من أن العمود موجود
if column_name in df.columns:
    # حساب القيم في الربيع الأول (5%) والربيع الثالث (95%) للتحديد
    lower_cap = df[column_name].quantile(0.05)
    upper_cap = df[column_name].quantile(0.95)

    # نسخ البيانات الأصلية لإجراء التعديل
    df_capped = df.copy()

    # تطبيق تحديد القيم الدنيا والعليا (Clipping) على العمود
    df_capped[column_name] = df_capped[column_name].clip(lower_cap, upper_cap)

    # عرض أول 15 صف من البيانات المعدلة
    print(df_capped.head(15))
else:
    print(f"العمود '{column_name}' غير موجود في البيانات.")
```
